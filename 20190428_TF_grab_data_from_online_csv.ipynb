Now that we covered how to grab some data stored in csv form from the web it might be nice to do something with it. 

```
import os
import matplotlib.pyplot as plt

import tensorflow as tf

print("TensorFlow version: {}".format(tf.__version__))
print("Eager execution: {}".format(tf.executing_eagerly()))

train_dataset_url = "https://raw.githubusercontent.com/openelections/openelections-data-co/master/2018/20181106__co__general__douglas__precinct.csv"

train_dataset_fp = tf.keras.utils.get_file(fname=os.path.basename(train_dataset_url),
                                           origin=train_dataset_url)

print("Local copy of the dataset file: {}".format(train_dataset_fp))
```

You might have noticed that header row had some content in it. We probably need to set the column names based on that header row so that we have things set up right.  

```
# column order in CSV file
column_names = ['county', 'precient', 'office', 'district', 'candidate', 'votes']

feature_names = column_names[:-1]
label_name = column_names[-1]

print("Features: {}".format(feature_names))
print("Label: {}".format(label_name))
```

This dataset is tiny... by machine leanring standards it is super small. That makes it easy to grab real quick as an exmaple, but it makes it less useful to using for training.

```
batch_size = 32

train_dataset = tf.data.experimental.make_csv_dataset(
    train_dataset_fp,
    batch_size, 
    column_names=column_names,
    label_name=label_name,
    num_epochs=1)
```
